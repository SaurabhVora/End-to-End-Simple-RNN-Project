```python
import numpy as np
import tensorflow
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent = ['the glass of milk',
        'the glass of juice',
        'the cup of tea',
        'I am a good boy',
        'I am a good developer',
        'understand the meaning of words',
        'your videos are good']
```

```python
sent
```

```python
voc_size = 10000
```

```python
one_hot_vector=[one_hot(words,voc_size) for words in sent]
one_hot_vector
```

```python
sent_length = 8
embedded_doc = pad_sequences(one_hot_vector,padding = 'pre',maxlen = sent_length)
print(embedded_doc)
```

```python
dim = 10
model = Sequential()
model.add(Embedding(voc_size,dim))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_doc)
```
